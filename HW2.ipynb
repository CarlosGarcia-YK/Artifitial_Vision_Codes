```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

def apply_morph_operations(image, kernel_size=(5, 5)):
    
    kernel = np.ones(kernel_size, np.uint8)
    dilation = cv2.dilate(image, kernel, iterations=1)
    erosion = cv2.erode(image, kernel, iterations=1)
    opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
    closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
    

    image_path = "pictures\microsoft.jpeg"
    image2 = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
  
    plt.figure(figsize=(10, 8))

    plt.subplot(231), plt.imshow(image2, cmap='gray'), plt.title('Original Image')
    
    plt.subplot(232), plt.imshow(dilation, cmap='gray'), plt.title('Dilation')
    plt.subplot(233), plt.imshow(erosion, cmap='gray'), plt.title('Erosion')
    plt.subplot(234), plt.imshow(opening, cmap='gray'), plt.title('Opening')
    plt.subplot(235), plt.imshow(closing, cmap='gray'), plt.title('Closing')

    plt.tight_layout()
    plt.show()
```

```python

def load_user_image(path):
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        print("Error")
        return None

    _, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
    return binary_image
image_path = "pictures\microsoft.jpeg"


user_image = load_user_image(image_path)

if user_image is not None:
    apply_morph_operations(user_image)
```
